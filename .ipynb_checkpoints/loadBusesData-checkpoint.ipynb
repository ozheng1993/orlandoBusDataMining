```python
import requests
import numpy as np
import math
import os
from sys import argv
import shutil
import datetime
import csv
import json
import pandas as pd
from pandas.io.json import json_normalize
from datetime import datetime
import time
from IPython.display import clear_output

from pprint import pprint
url = 'http://golynx.doublemap.com/map/v2/buses'
#create folder
if os.path.exists('data'):
    print("file data exist")
else:
    print("file data not exist,created")
    os.mkdir('data')
if os.path.exists('data/bus'):
    print("file data/bus exist")
else:
    os.mkdir('data/bus')
    print("file data/bus not exist,created")

csvKeys=["index",
        
        'direction', 
        'headsign', 
        'id', 
        'loc.heading',
        'loc.lat',
        'loc.lon',
        'loc.source',
        'loc.speed',
        'loc.time',
        'loc.routeId',
        'routeShortName', 
        'vehicleType'
    ]
csvValuse=[None, None,None,None,None,None,None,None,None,None,None,None,]
if os.path.exists('data/bus/busRaw.csv'):
    print("file data/bus/busRaw.csv exist")
else:
    with open('data/bus/busRaw.csv', 'wb') as outcsv:
        writer = csv.DictWriter(outcsv, fieldnames = csvKeys)
        writer.writeheader()
    outcsv.close()
    print("file data/busRaw.csv not exist,created")
counter=0;
start_time=datetime.now()
while True:
    counter+=55;
    
    resp = requests.get(url=url)
    try:
        data = resp.json()
    except:
        print "This is an error message!"
        continue
    data = resp.json()
    data2=json_normalize(data['vehicles'])
    with open('data/bus/dataRaw.txt', 'w') as outfile:
        json.dump(data, outfile)
    #print(data2)
    ##pprint(data)
    df = pd.DataFrame(data2, columns = [
        'direction', 
        'headsign', 
        'id', 
        'loc.heading',
        'loc.lat',
        'loc.lon',
        'loc.source',
        'loc.speed',
        'loc.time',
        'routeId',
        'routeShortName', 
        'vehicleType'
    ]) 
    
    print ("running time : "+str(start_time))
    print ("data collected : "+str(counter))
    print df
    df.to_csv('data/bus/busRaw.csv', mode='a', header=False)
    time.sleep(1)
    clear_output(wait=True)
```
